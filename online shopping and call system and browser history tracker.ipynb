```python
# ---------------- Task 1: Online Shopping Cart System ----------------
print("\n=== Task 1: Online Shopping Cart System (List) ===")

# Initialize empty shopping cart
cart = []

def add_item(item):
    cart.append(item)
    print(f"🛒 '{item}' added to cart.")

def remove_item(item):
    if item in cart:
        cart.remove(item)
        print(f"'{item}' removed from cart.")
    else:
        print("Item not found in cart.")

def view_cart():
    if not cart:
        print("🛍️ Your cart is empty.")
    else:
        print("\nItems in your cart:")
        for i, item in enumerate(cart, start=1):
            print(f"{i}. {item}")
        print("Total items:", len(cart))

# Simulate actions
add_item("Laptop")
add_item("Headphones")
add_item("Mouse")
view_cart()
remove_item("Mouse")
view_cart()
```

```python
# ---------------- Task 2: Customer Service Call Center ----------------
print("\n=== Task 2: Customer Service Call Center (Queue) ===")

from collections import deque

# Initialize queue
call_queue = deque()

def enqueue_call(caller):
    call_queue.append(caller)
    print(f"📞 New call added: {caller}")

def dequeue_call():
    if call_queue:
        attended = call_queue.popleft()
        print(f" Attending call from: {attended}")
    else:
        print("No calls waiting.")

def display_calls():
    if call_queue:
        print("\nWaiting calls:")
        for caller in call_queue:
            print("-", caller)
    else:
        print("No calls in the queue.")

# Simulate actions
enqueue_call("Ali")
enqueue_call("Sana")
enqueue_call("Ahmed")
display_calls()
dequeue_call()
display_calls()
```

```python
# ---------------- Task 3: Browser History Tracker ----------------
print("\n=== Task 3: Browser History Tracker (Stack) ===")

# Initialize browser history stack
history_stack = []

def visit_page(url):
    history_stack.append(url)
    print(f"🌐 Visited: {url}")

def go_back():
    if history_stack:
        last_page = history_stack.pop()
        print(f"⬅️ Going back from: {last_page}")
    else:
        print("No history available.")

def current_page():
    if history_stack:
        print("📍 Current page:", history_stack[-1])
    else:
        print("No current page.")

def show_history():
    if history_stack:
        print("\nAll visited URLs (most recent first):")
        for url in reversed(history_stack):
            print("-", url)
    else:
        print("No browsing history.")

# Simulate actions
visit_page("google.com")
visit_page("youtube.com")
visit_page("github.com")
current_page()
show_history()
go_back()
current_page()
show_history()
```
